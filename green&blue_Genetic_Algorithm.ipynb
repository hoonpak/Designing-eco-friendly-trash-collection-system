# Chromosome

* <H4> 1. 최소의 capacity를 채우지 못한 경우에 주변에 쓰래기가 있는 경로가 없을 경우 depot으로 되돌아오지 못하게함</H4>

* <H4> 2. 그 다음 경로로 움직일 때 차량의 수거량이 해당 쓰래기의 수거 가능 최대 용량 보다 크기 직전에 최소 경로로 복귀</H4>

* <H4> 3. 만약 전체 경로에 쓰래기가 없다면 바로 최소 경로로 복귀</H4>

* <H4> 4. depot에 도착하면 알고리즘 종료</H4>

<h6> 쓰래기 별 최소 경로 만큼만 수거해줌 - 회색 : 5, 초록 : 6, 파랑 : 5 </h6>

```python
#gene : 차량 1개의 경로 후보 1개
def course(capacity, car_path_li=trash_car_path, shortest_path_to_zero=shortest_course_to_zero):
    trash=copy.deepcopy(trash_li)
    chromosome=[]
    
    while True:
        
        if sum([sum(trash[t])for t in range(len(trash))]) == 0:
                break
                
        gene=[0]; capa=0; a=0
        
        while True:
            
            b_choice_list = []
            
            for i in car_path_li[a]:
                if trash[a][i] != 0:
                    b_choice_list.append(i)
            
            if b_choice_list == []:
                b_choice_list = car_path_li[a]
            
            b = random.choice(b_choice_list)
                
            if capa + trash[a][b] > capacity:
                for i in range(len(shortest_path_to_zero[b])):
                    gene.append(shortest_path_to_zero[b][i])
                break
            
            gene.append(b)
            
            capa+=trash[a][b]
            
            trash[a][b]=0
            if sum([sum(trash[t])for t in range(len(trash))]) == 0:
                for x in range(len(shortest_path_to_zero[b])-1):
                    gene.append(shortest_path_to_zero[b][x+1])
                break
                
            a = b
            
            if gene[-1]==0:
                break
        
        chromosome.append(gene)
    return chromosome
```

```python
def chromosome():
    green_course = course(green_capacity)
    trash=copy.deepcopy(trash_li)
    blue_course=[]
    
    for i in range(len(green_course)//3):
        blue_path = [[] for i in range(N)]
        if i == (len(green_course)//3):
            blue_path_candidate=sum(green_course[i*3:],[])
            for j in range(len(blue_path_candidate)-1):
                blue_path[blue_path_candidate[j]].append(blue_path_candidate[j+1])
                
        else:
            blue_path_candidate=sum(green_course[i*3:i*3+3],[])
            for j in range(len(blue_path_candidate)-1):
                blue_path[blue_path_candidate[j]].append(blue_path_candidate[j+1])
        
        blue_path = [list(set(k)) for k in blue_path]
        blue_path[0].remove(0)
        
        while True:
        
            if sum([sum(trash[t])for t in range(len(trash))]) == 0:
                    break

            gene=[0]; capa=0; a=0

            while True:

                b_choice_list = []
                
                for m in blue_path[a]:
                    if trash[a][i] != 0:
                        b_choice_list.append(i)
                
                if b_choice_list == []:
                    for m in trash_car_path[a]:
                        if trash[a][i] != 0:
                            b_choice_list.append(i)
                            
                    if b_choice_list == []:
                        b_choice_list = trash_car_path[a]

                b = random.choice(b_choice_list)

                if capa + trash[a][b] > blue_capacity:
                    for n in range(len(shortest_course_to_zero[b])):
                        gene.append(shortest_course_to_zero[b][n])
                    break

                gene.append(b)

                capa+=trash[a][b]

                trash[a][b]=0
                if sum([sum(trash[t])for t in range(len(trash))]) == 0:
                    for x in range(len(shortest_course_to_zero[b])-1):
                        gene.append(shortest_course_to_zero[b][x+1])
                    break

                a = b

                if gene[-1]==0:
                    break

            blue_course.append(gene)
        
        if sum([sum(trash[t])for t in range(len(trash))]) == 0:
            break
            
    
    return green_course, blue_course
```

# Fitness Function

  * <H3> 쓰래기를 전부 수거하지 못하는 경우 매우 큰 값을 return하게 만들어 줌 </H3>
  * <H3> 하지만 쓰래기가 많이 수거된 것 일수록 더 좋은 값을 갖게함. </H3>

*****

```python
def one_course_fitness(chromosome, capacity):
    
    trash = copy.deepcopy(trash_li)
    
    for chro in chromosome:
        capa=0
        for i in range(len(chro)-1):
            a = chro[i]
            b = chro[i+1]
            capa+=trash[a][b]
            if capa > capacity:
                break
            trash[a][b] = 0
            
    if sum([sum(trash[t])for t in range(len(trash))]) != 0:
        trash_remain_amount = sum([sum(trash[t])for t in range(len(trash))])
        travel_time = sum([len(c)-1 for c in chromosome])*5
        return trash_remain_amount*70 + travel_time*10 + 308+1000000000
    else:
        trash_total_amount = 308
    
    travel_time = sum([len(c)-1 for c in chromosome])*5
    trash_collect_time = trash_total_amount*1
    
    return travel_time + trash_collect_time
```

```python
def fitness_function(chromosome):
    green_chromosome = chromosome[0]
    blue_chromosome = chromosome[1]
    return one_course_fitness(green_chromosome, green_capacity) + one_course_fitness(blue_chromosome, blue_capacity)
```

# Selection

```python
def selection(population):
    
    population_size = len(population)
    select_list=[]
    
    population_fitness = [fitness_function(i) for i in population]
    population_fit_reciprocal = [ 1/i for i in population_fitness]
    population_fit_pro = [i/sum(population_fit_reciprocal) for i in population_fit_reciprocal]
    population_fit_pro_sort = sorted(population_fit_pro, reverse=True)
    chromosome_index_list=[population_fit_pro.index(i) for i in population_fit_pro_sort]
    prop = 0
    pro_cumulative_sum = [0]
    for i in population_fit_pro_sort:
        prop += i
        pro_cumulative_sum.append(prop)
        
    select_pro = [random.random() for i in range(population_size)]
    select_pro = sorted(select_pro)
    
    for i in range(len(pro_cumulative_sum)-1):
        for j in select_pro:
            if pro_cumulative_sum[i] < j <= pro_cumulative_sum[i+1]:
                select_index = chromosome_index_list[i]
                select_list.append(population[select_index])
                
    return select_list
```

```python
def cross_over(population):
    cross_over_population = copy.deepcopy(population)
    green_population = [i[0] for i in population]
    blue_population = [i[1] for i in population]

    population_size = len(population)
    
    crossover_pro = [random.random() for i in range(population_size)]
    
    for i in range(population_size):
        if crossover_pro[i] > 0.5:
            target_chromosome=population[i][0]
            crossover_gene_index_1=np.random.randint(0, len(target_chromosome))
            crossover_gene_value=random.choice(target_chromosome[crossover_gene_index_1])
            crossover_gene_value_index_1=target_chromosome[crossover_gene_index_1].index(crossover_gene_value)
            new_genes=[gene for gene in target_chromosome if gene != target_chromosome[crossover_gene_index_1]]
            
            for j in new_genes:
                if crossover_gene_value in j:
                    crossover_gene_value_index_2=j.index(crossover_gene_value)
                    crossover_gene_index_2=target_chromosome.index(j)
                    
                    cross_gene_1=target_chromosome[crossover_gene_index_1][crossover_gene_value_index_1:]
                    cross_gene_2=target_chromosome[crossover_gene_index_2][crossover_gene_value_index_2:]
                    green_population[i][crossover_gene_index_1][crossover_gene_value_index_1:]=cross_gene_2
                    green_population[i][crossover_gene_index_2][crossover_gene_value_index_2:]=cross_gene_1
                    break
        
            green_course = green_population[i]
            trash=copy.deepcopy(trash_li)
            blue_course=[]

            for i_ in range(len(green_course)//3):
                blue_path = [[] for i in range(N)]
                if i_ == (len(green_course)//3):
                    blue_path_candidate=sum(green_course[i_*3:],[])
                    for j_ in range(len(blue_path_candidate)-1):
                        blue_path[blue_path_candidate[j_]].append(blue_path_candidate[j_+1])

                else:
                    blue_path_candidate=sum(green_course[i_*3:i_*3+3],[])
                    for j_ in range(len(blue_path_candidate)-1):
                        blue_path[blue_path_candidate[j_]].append(blue_path_candidate[j_+1])

                blue_path = [list(set(k)) for k in blue_path]
                blue_path[0].remove(0)

                while True:

                    if sum([sum(trash[t])for t in range(len(trash))]) == 0:
                            break

                    gene=[0]; capa=0; a=0

                    while True:

                        b_choice_list = []

                        for m in blue_path[a]:
                            if trash[a][i_] != 0:
                                b_choice_list.append(i_)

                        if b_choice_list == []:
                            for m in trash_car_path[a]:
                                if trash[a][i_] != 0:
                                    b_choice_list.append(i_)

                            if b_choice_list == []:
                                b_choice_list = trash_car_path[a]

                        b = random.choice(b_choice_list)

                        if capa + trash[a][b] > blue_capacity:
                            for n in range(len(shortest_course_to_zero[b])):
                                gene.append(shortest_course_to_zero[b][n])
                            break

                        gene.append(b)

                        capa+=trash[a][b]

                        trash[a][b]=0
                        if sum([sum(trash[t])for t in range(len(trash))]) == 0:
                            for x in range(len(shortest_course_to_zero[b])-1):
                                gene.append(shortest_course_to_zero[b][x+1])
                            break

                        a = b

                        if gene[-1]==0:
                            break

                    blue_course.append(gene)

                if sum([sum(trash[t])for t in range(len(trash))]) == 0:
                    break


            blue_population[i] = blue_course
            cross_over_population[i] = (green_population[i], blue_population[i])
        
    cross_over_population_fitness = [fitness_function(i) for i in cross_over_population]
    cross_over_population_fitness_sort = sorted(cross_over_population_fitness)
    chromosome_index_list=[cross_over_population_fitness.index(i) for i in cross_over_population_fitness_sort]
    result_list=[cross_over_population[i] for i in chromosome_index_list]
    
    return result_list
```

```python
def route(smallest_capacity=gray_smallest_capacity,capacity=gray_capacity):
    trash = copy.deepcopy(trash_li)
    route=[0]
    a = 0
    capa = 0

    while True:

        b_choice_list = []

        if capa <= smallest_capacity:
            for i in trash_car_path[a]:
                if trash[a][i] != 0:
                    b_choice_list.append(i)
            if b_choice_list == []:
                b_choice_list = trash_car_path_not_zero[a]

        else :
            for i in trash_car_path[a]:
                if trash[a][i] != 0:
                    b_choice_list.append(i)
            if b_choice_list == []:
                b_choice_list = trash_car_path[a]

        b = random.choice(b_choice_list)

        if capa + trash[a][b] > capacity:
            for i in shortest_course_to_zero[b]:
                route.append(i)
            break

        capa += trash[a][b]

        trash[a][b] = 0

        if sum([sum(trash[t])for t in range(len(trash))]) == 0:
            for i in shortest_course_to_zero[b]:
                route.append(i)
            break

        route.append(b)

        if b == 0:
            break    

        a = b
    return route
```

```python
def mutation(population):
    mutation_population = copy.deepcopy(population)
    green_population = [i[0] for i in population]
    blue_population = [i[1] for i in population]    
    
    N = len(population)
    
    mutation_pro = [random.random() for i in range(N)]
    
    for i in range(N):
        if mutation_pro[i] > 0.7:
            mutation_gene = np.random.randint(0, len(green_population[i]))
            green_population[i][mutation_gene] = route()
            
            green_course = green_population[i]
            trash=copy.deepcopy(trash_li)
            blue_course=[]

            for i_ in range(len(green_course)//3):
                blue_path = [[] for i in range(N)]
                if i_ == (len(green_course)//3):
                    blue_path_candidate=sum(green_course[i_*3:],[])
                    for j_ in range(len(blue_path_candidate)-1):
                        blue_path[blue_path_candidate[j_]].append(blue_path_candidate[j_+1])

                else:
                    blue_path_candidate=sum(green_course[i_*3:i_*3+3],[])
                    for j_ in range(len(blue_path_candidate)-1):
                        blue_path[blue_path_candidate[j_]].append(blue_path_candidate[j_+1])

                blue_path = [list(set(k)) for k in blue_path]
                blue_path[0].remove(0)

                while True:

                    if sum([sum(trash[t])for t in range(len(trash))]) == 0:
                            break

                    gene=[0]; capa=0; a=0

                    while True:

                        b_choice_list = []

                        for m in blue_path[a]:
                            if trash[a][i_] != 0:
                                b_choice_list.append(i_)

                        if b_choice_list == []:
                            for m in trash_car_path[a]:
                                if trash[a][i_] != 0:
                                    b_choice_list.append(i_)

                            if b_choice_list == []:
                                b_choice_list = trash_car_path[a]

                        b = random.choice(b_choice_list)

                        if capa + trash[a][b] > blue_capacity:
                            for n in range(len(shortest_course_to_zero[b])):
                                gene.append(shortest_course_to_zero[b][n])
                            break

                        gene.append(b)

                        capa+=trash[a][b]

                        trash[a][b]=0
                        if sum([sum(trash[t])for t in range(len(trash))]) == 0:
                            for x in range(len(shortest_course_to_zero[b])-1):
                                gene.append(shortest_course_to_zero[b][x+1])
                            break

                        a = b

                        if gene[-1]==0:
                            break

                    blue_course.append(gene)

                if sum([sum(trash[t])for t in range(len(trash))]) == 0:
                    break


            blue_population[i] = blue_course
            mutation_population[i] = (green_population[i], blue_population[i])
            
        else:
            pass
            
    mutation_population_fitness = [fitness_function(i) for i in mutation_population]
    mutation_population_fitness_sort = sorted(mutation_population_fitness)
    chromosome_index_list=[mutation_population_fitness.index(i) for i in mutation_population_fitness_sort]
    result_list=[mutation_population[i] for i in chromosome_index_list]
    
    return result_list
```
